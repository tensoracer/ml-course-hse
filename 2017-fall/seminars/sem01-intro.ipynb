#### 2. Напоминание материала лекции

Вспомним основные виды задач машинного обучения, о которых говорилось на лекции, и обсудим несколько других.

- Задача регрессии — задача с вещественной целевой переменной, где основная цель предсказать значение близкое к фактическому в смысле некоторого функционала ошибки. Простой пример — предсказание цены дома по его параметрам.
- Задача классификации (в том числе многоклассовая и многоклассовая с пересекающимися классами) — задача, где целевая переменная принимает значение из некоторого фиксированного конечного множества значений (или несколько в случае пересекающихся классов). Например, задача тегирования видеороликов — многоклассовая классификация с пересекающимися классами.
- Задача кластеризации — задача, в которой выборка не имеет какой-то целевой переменной, а необходимо разделить объекты на группым по некоторым свойствам. На практике нередко сложно оценить качество при решении этой задачи. Пример — разделить пользователей сервиса на группы для дальнейшего исследования клиентов. 
- Задача поиска выбросов — задача, похожая на бинарную классификацию с преобладанием объектов одного класса с тем отличием, что объекты "редкого" класса могут быть не похожими между собой, а лишь отличаться от объектов основного класса. Кроме этого, на практике объектов-выбросов в выборке может и не быть. Например, задача поиска фальшивых монет. Если быть строже, задачи этого класса можно разделить на детектирование выбросов и новизны.
- Задача оценивания плотности — задача приближения распределения объектов. Иногда на практике необходимо оценивать "типичность" объектов, что позволяет обнаруживать нестандартное поведение или события. Этим задача часто похожа на поиск выбросов и может решаться схожими способами. 
- Задача ранжирования — задача, в которой необходимо отсортировать объекты выборки по некоторому критерию (релеватности). При этом иногда нужно сортировать все объекты глобально, а иногда только внутри некоторых групп (например, в поиске необходимо сортировать ссылки для каждого запроса отдельно). Типичный пример — поисковая система.
- Задача обучения с подкреплением — задача, в которой обучающая система учится воздействовать со средой, не имея сведений о ней. Например, двуногий робот учится управлять своими шарнирами, двигаясь по ровной поверхности. Такие задачи достаточно специфичны и не будут рассматриваться в этом курсе.

Иногда выделяют и другие более редкие виды задач. Некоторые из них могут пересекаться с описанными выше. Например, задача построения рекомендаций может решаться как задача бинарной классификации или как задача ранжирования.

Теперь откроем несколько таблиц с данными. Делается это буквально в одну строчку. А данные у нас с прошлогоднего конкурса Sberbank Data Science Journey. Представляют собой транзакции клиентов, использующих банковские карты. Каждая строка соответствует одной транзакции и включает в себя дату и время, mcc код (категория товара или услуги), тип транзакции (как и на чём совершена), сумма и некий код транзакции.

```python
transactions = pd.read_csv('data/transactions.csv')

transactions.head()
```

Как видим, конкретной даты у нас нет. Только номер дня и время. 

Также у нас есть описание всех mcc кодов.

```python
mcc = pd.read_csv('data/tr_mcc_codes.csv', sep=';')

mcc.head()
```

Настоящую дату мы не знаем, но давайте вычислим. Ведь как бы организаторы конкурсов не пытались скрыть от нас информацию, датасайентист всё равно узнает правду. Для начала извлечём номер дня из второй колонки (заметим, как легко произодить манипуляции со значениями всего столбца). 

```python
transactions['day'] = transactions.tr_datetime.apply(lambda x: int(x.split(' ')[0]))
```

Давайте посчитаем, сколько транзакций совершается в каждый из дней. Для этого сгруппируем строки по колонке с номером дня и посчитаем количество. Всё это через методы библиотеки Pandas. Изобразим на графике.

```python
gr = transactions.groupby('day')['amount'].count()

plt.figure(figsize=(10, 5))
plt.plot(gr.index, gr)
plt.xlabel('day')
plt.ylabel('count')
plt.show()
```

И что мы видим? В какой же период код количество транзакций на некоторое время заметно уменьшается? Правильно, на Новый год. Найдём этот день и отцентрируем данные, что 0-й день был 1 января.

```python
np.argmin(gr)
```

```python
transactions.day -= 153
```

С этим разобрались, давайте смотреть покрупнее.

```python
part_transaction = transactions.loc[np.logical_and(transactions.day >= 10, transactions.day <= 63)]

gr = part_transaction.groupby('day')['amount'].count()

plt.figure(figsize=(10, 5))
plt.plot(gr.index, gr, 'o-')
plt.plot([31, 31], [10000, 15000])
plt.xlabel('day')
plt.ylabel('count')
plt.show()
```

Что мы можем сказать на этот раз? Видим, что есть циклы длиной в 7 дней, то есть видны недельные перепады. В какой день люди меньше используют карты (сидят дома, не закупают продукты, не обедают)? Пожалуй, в воскресенье. А на графике 31-й день (а так как отсчёт был с 0, то 32-й) года является выходным днём. Посмотрев календарь, видим, что этот был либо 2009, либо 2015 год (более ранний слишком маловероятен для конкурса 2016 года).

Как мы можем различить 2009 и 2015? Возможно, по праздникам. Сравнив календари этих двух лет, увидим, что 4 мая в 2009 году был рабочим днём, а в 2015 выходным. Изобразим этот период на графике.

```python
part_transaction = transactions.loc[np.logical_and(transactions.day >= 31+28+31+30, transactions.day <= 31+28+31+30+15)]

gr = part_transaction.groupby('day')['amount'].count()

plt.figure(figsize=(10, 5))
plt.plot(gr.index, gr, 'o-')
plt.xlabel('day')
plt.ylabel('count')
plt.show()
```

То есть четвёртое мая было выходным днём! Значит, это 2015 год, что вполне стыкуется с тем, что конкурс проходил в 2016 году и данные были достаточно свежими.

Также мы могли анализировать снятия денег в банкомате, ведь люди часто делают это в банках, которые в выходные дни работат довольно редко. Или же посмотреть на количество обедов в рабочие часы.
